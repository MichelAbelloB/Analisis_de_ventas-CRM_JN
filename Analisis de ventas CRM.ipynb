### Paso 4. Análisis de la relación entre variables

```python
#Realizamos la matriz de correlación entre las variables
correlation_matrix = data.corr(method ='pearson')

# Crear un mapa de calor con seaborn
plt.figure(figsize=(15,10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlación (Valores por encima del umbral)')
plt.show()
```

**15**. ¿Que información se desprende de una matriz de correlación?

**Respuesta:**

Una matriz de correlación proporciona información sobre la relación lineal entre pares de variables en un conjunto de datos. Los valores en la matriz varían entre -1 y 1.

- Si el valor es cercano a 1, indica una correlación positiva fuerte, lo que significa que cuando una variable aumenta, la otra también tiende a aumentar.
- Si el valor es cercano a -1, indica una correlación negativa fuerte, lo que significa que cuando una variable aumenta, la otra tiende a disminuir.
- Si el valor es cercano a 0, indica una correlación débil o nula, lo que significa que no hay una relación lineal clara entre las variables.

La matriz de correlación es útil para identificar patrones en los datos, entender cómo las variables están relacionadas entre sí y seleccionar variables para análisis posteriores o modelos predictivos. Sin embargo, es importante recordar que la correlación no implica causalidad, es decir, el hecho de que dos variables estén correlacionadas no significa necesariamente que una cause la otra.

**17**. Indicad que implicaría para el análisis que las variables estén altamente correlacionadas.

**Respuesta:**

Cuando las variables en un conjunto de datos están altamente correlacionadas, esto puede complicar el análisis de varias maneras. En primer lugar, la redundancia de información puede surgir, lo que significa que incluir todas las variables correlacionadas puede aumentar la complejidad sin mejorar la capacidad predictiva o explicativa del modelo. Además, la multicolinealidad puede dificultar la interpretación de los coeficientes de regresión y hacer que las estimaciones sean menos estables. Esta situación también puede sesgar la interpretación de la importancia de las variables, lo que lleva a conclusiones erróneas en el análisis de datos. Por último, los modelos construidos con variables altamente correlacionadas pueden ser más sensibles a cambios en los datos, lo que puede afectar su robustez y generalización. En resumen, la alta correlación entre variables puede complicar el análisis y afectar la precisión y validez de los resultados obtenidos.
